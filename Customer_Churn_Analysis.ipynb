**Converted 0 and 1 values of senior citizen into Yes or No make it easier to understand**

```python
def conv(value):
  if value == 1:
    return "Yes"
  else:
      return "No"
df['SeniorCitizen'] = df['SeniorCitizen'].apply(conv)
```

```python
ax = sns.countplot(x='Churn', data=df)
ax.bar_label(ax.containers[0])
plt.show()
```

```python
plt.figure(figsize=(5,4))
gb=df.groupby("Churn").agg({"Churn":"count"})
plt.pie(gb['Churn'],labels=gb.index,autopct='%1.1f%%')
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.countplot(x='gender',hue='Churn',data=df)
plt.title('Churn By Gender',fontsize=10)
plt.show()
```

```python
# Group by 'gender' and 'Churn', and count
grouped = df.groupby(['gender', 'Churn']).size().unstack(fill_value=0)

# Calculate percentages
percentages = grouped.div(grouped.sum(axis=1), axis=0) * 100

# Plot stacked bar chart
ax = percentages.plot(kind='bar', stacked=True, figsize=(6,4), colormap='Pastel1')

# Add percentage labels
for container in ax.containers:
    labels = [f'{v:.1f}%' if v > 0 else '' for v in container.datavalues]
    ax.bar_label(container, labels=labels, label_type='center', fontsize=8)

plt.title('Churn By Gender (in %)')
plt.ylabel('Percentage')
plt.xlabel('Gender')
plt.legend(title='Churn', loc='upper right')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(x="tenure", data=df, bins=72, hue='Churn', multiple="stack")
plt.title("Customer Tenure Distribution by Churn")
plt.xlabel("Tenure (Months)")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```
